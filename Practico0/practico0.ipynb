**Histograma:**  
Un histograma representa la distribuci√≥n de una variable num√©rica para uno o varios grupos. Los valores se dividen en intervalos (*bins*), y cada intervalo se representa como una barra.  
üîó [M√°s informaci√≥n](https://python-graph-gallery.com/histogram/)



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style="darkgrid")


df = sns.load_dataset("iris")

sns.displot(df["sepal_length"],
            kde=False,
            rug=False)
plt.show()
```

**Boxplot (‚Äúdiagrama de caja y l√≠nea‚Äù):**  
Un *boxplot* resume la distribuci√≥n de una variable num√©rica para uno o varios grupos. Permite obtener r√°pidamente la mediana, los cuartiles y los valores at√≠picos, pero oculta los puntos de datos individuales del conjunto.  
üîó [M√°s informaci√≥n](https://python-graph-gallery.com/boxplot/)



```python

sns.set_theme(style="darkgrid")
df = sns.load_dataset('iris')
 
ax = sns.boxplot(x="species", y="sepal_length", data=df)

medians = df.groupby(['species'])['sepal_length'].median().values
nobs = df['species'].value_counts().values
nobs = [str(x) for x in nobs.tolist()]
nobs = ["n: " + i for i in nobs]
 

pos = range(len(nobs))
for tick,label in zip(pos,ax.get_xticklabels()):
    ax.text(pos[tick],
            medians[tick] + 0.03,
            nobs[tick],
            horizontalalignment='center',
            size='x-small',
            color='w',
            weight='semibold')
 
plt.show()
```

**Diagrama lineal:**  
Un diagrama lineal muestra la evoluci√≥n de una o varias variables num√©ricas continuas.  
üîó [M√°s informaci√≥n](https://python-graph-gallery.com/line-chart/)



```python

df = pd.DataFrame({
    'x_values': range(1,11),
    'y1_values': np.random.randn(10),
    'y2_values': np.random.randn(10)+range(1,11),
    'y3_values': np.random.randn(10)+range(11,21)
    }
)

plt.plot(
    'x_values', 'y1_values', data=df,
    marker='o', 
    markerfacecolor='blue',
    markersize=8, 
    color='skyblue',
    linewidth=4 
)

plt.plot(
    'x_values', 'y2_values', data=df,
    marker='', 
    color='olive',
    linewidth=2 
)

plt.plot(
    'x_values', 'y3_values', data=df,
    marker='', 
    color='darkred', 
    linewidth=3, 
    linestyle='dashed', 
    label="toto" 
)

plt.legend()

plt.show()
```

**Series temporales:**  
Los gr√°ficos de series temporales incluyen todos los gr√°ficos que representan la evoluci√≥n de un valor num√©rico. Gr√°ficos de l√≠neas, *streamgraph*, diagramas de barras y gr√°ficos de √°rea se utilizan para la visualizaci√≥n de series temporales.  
üîó [M√°s informaci√≥n](https://python-graph-gallery.com/timeseries/)



```python

my_count = ["France", "USA", "Japan", "Brazil"]


df = pd.DataFrame({
    "country": np.repeat(my_count, 10),
    "years": list(range(2000, 2010)) * 4,
    "value": np.random.rand(40)
})


g = sns.FacetGrid(df, col='country', hue='country', col_wrap=2) 


g = g.map(plt.plot, 'years', 'value')
g = g.map(plt.fill_between, 'years', 'value', alpha=0.2).set_titles("{col_name}")

plt.subplots_adjust(top=0.85)
g.fig.suptitle('Evoluci√≥n del valor en 4 pa√≠ses')


plt.show()
```

**Diagrama de coordenadas paralelas (o ‚Äúdiagrama paralelo‚Äù):**  
Un diagrama paralelo permite comparar las caracter√≠sticas de varias observaciones individuales (series) en un conjunto de variables num√©ricas.  
üîó [M√°s informaci√≥n](https://python-graph-gallery.com/parallel-plot/)



```python
from pandas.plotting import parallel_coordinates

data = sns.load_dataset('iris')

parallel_coordinates(data, 'species', colormap=plt.get_cmap("Set2"))
plt.show()
```

**Gr√°fico de dispersi√≥n (*Scatter Plot*):**  
Un gr√°fico de dispersi√≥n muestra la relaci√≥n entre dos variables num√©ricas, una representada en el eje X (horizontal) y la otra en el eje Y (vertical). Cada punto de datos se representa como un c√≠rculo.  
üîó [M√°s informaci√≥n](https://python-graph-gallery.com/scatter-plot/)

```python

df = sns.load_dataset('iris')
sns.scatterplot(x=df["petal_length"], y=df["petal_width"], hue=df["species"])

```


**Gr√°ficos de l√≠neas con bandas de error (Errorband Lineplot):**  
Es un gr√°fico de l√≠neas que muestra la evoluci√≥n de una variable num√©rica a lo largo del tiempo o en diferentes condiciones, con una banda de error sombreada alrededor de la l√≠nea principal. Esta banda representa la incertidumbre o la variabilidad en los datos, como el intervalo de confianza o la desviaci√≥n est√°ndar.
Se usan cuando hay m√∫ltiples mediciones por cada punto en el eje X y sea relevante mostrar la variabilidad de los datos.

```python

sns.set_theme(style="darkgrid")

fmri = sns.load_dataset("fmri")

sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)

```
